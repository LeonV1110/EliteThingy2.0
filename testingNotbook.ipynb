# Create tables

```python
import sqlite3

connection = sqlite3.connect('smollEDDN')
cursor = connection.cursor()

cursor.execute("DROP TABLE IF EXISTS ExoBio")

exobio_table = '''CREATE TABLE ExoBio (
        Name VARCHAR (255) NOT NULL,
        TimeStamp VARCHAR (255) NOT NULL,
        System VARCHAR (255) NOT NULL,
        SystemAddress VARCHAR (255),
        BodyName VARCHAR (255) NOT NULL,
        BodyID INT)
        ;'''

cursor.execute(exobio_table)

uniqueQuerry = '''CREATE UNIQUE INDEX Name_TimeStamp ON ExoBio (Name, TimeStamp)'''

cursor.execute(uniqueQuerry)

connection.commit()
connection.close()
```

```python
connection = sqlite3.connect('smollEDDN')
cursor = connection.cursor()

cursor.execute("DROP TABLE IF EXISTS Bodies")

bodies_table = '''CREATE TABLE Bodies (
        BodyName VARCHAR (255) NOT NULL,
        BodyID VARCHAR (255) NOT NULL,
        SystemAddress INT NOT NULL,
        StarSystem VARCHAR (255) NOT NULL,
        TimeStamp VARCHAR (255) NOT NULL,
        SurfaceTemperature FLOAT,
        SurfaceGravity FLOAT,
        ParentStarID INT,
        PlanetClass VARCHAR (255),
        AtmosphereType VARCHAR (255),
        SurfacePressure FLOAT,
        Volcanism VARCHAR (255)
        );'''

cursor.execute(bodies_table)

uniqueQuerry = '''CREATE UNIQUE INDEX Body_System_TimeStamp ON Bodies (BodyName,StarSystem, TimeStamp)'''

cursor.execute(uniqueQuerry)

connection.commit()
connection.close()
```

# Read Tables

```python
import sqlite3

connection = sqlite3.connect('smollEDDN')
cursor = connection.cursor()

querry = '''SELECT * FROM Bodies'''

cursor.execute(querry)

print("data")
data = cursor.fetchall()
for row in data:
    print(row)

connection.commit()
connection.close()
```
